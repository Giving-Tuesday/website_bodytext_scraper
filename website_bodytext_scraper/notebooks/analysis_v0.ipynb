```python
def calculate_metrics(df):
    total = len(df)
    success_rate = df['success'].mean() * 100  
    stripe_detected_rate = df['stripe_detected'].mean() * 100
    error_code_rate = df['error_code'].notna().mean() * 100
    
    # Display results
    metrics = {
        'Total Records': total,
        'Success Rate (%)': success_rate,
        'Stripe Detected Rate (%)': stripe_detected_rate,
        'Error Code Rate (%)': error_code_rate
    }
    
    return pd.DataFrame([metrics])
```

```python
metrics_df = calculate_metrics(df)
metrics_df
```
